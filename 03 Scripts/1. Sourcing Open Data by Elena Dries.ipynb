### 2.4. Mixed type columns

```python
# Checking for mixed-type data in 'core_met':

for column in core_met.columns.tolist():
  weird = (core_met[[column]].applymap(type) != core_met[[column]].iloc[0].apply(type)).any(axis = 1)
  if len (core_met[weird]) > 0:
    print (column)
```

No mixed type columns have been detected.
